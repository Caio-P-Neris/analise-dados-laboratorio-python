```python
import pandas as pd
import numpy
from google.colab import drive
import matplotlib.pyplot as plt
%matplotlib inline
drive.mount('/content/drive')
prata = pd.read_csv("/content/drive/MyDrive/Prata-Correta.csv", on_bad_lines='skip', sep=';')
nanica = pd.read_csv("/content/drive/MyDrive/NanicaFormatado.csv", on_bad_lines='skip', sep=';')
semban= pd.read_csv("/content/drive/MyDrive/SemBanana.csv", on_bad_lines='skip', sep=';')

x1 = prata['x']
y1 = prata['y']
x2 = nanica['x']
y2 = nanica['y']
x3 = semban['x']
y3 = semban['y']

plt.plot(x1,y1,'b',label='prata')
plt.plot(x2,y2,'r',label='nanica')
plt.plot(x3,y3,'g',label='amostra sem banana')

#plt.plot([x1, x2, x3], [y1, y2, y3])
plt.title('Comparação: Antes e depois da ação dos pedaços das cascas')
plt.xlabel('Comprimento de onda (nm)')
plt.ylabel('Absorbância')
plt.legend()
```
